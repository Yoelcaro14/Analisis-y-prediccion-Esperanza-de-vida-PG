HaY 270 datos faltantes<br>
Vamos a identificarlos

```python
#Vemos que paises tienen TODOS sus datos faltantes
print("Paises con TODOS datos nulos")
paises_nulos=[]
for i in paises:
    mask = tbl_contaminacion_aire["id_pais"] == i
    df_filtrada = tbl_contaminacion_aire[mask]
    a = df_filtrada.iloc[:,2].isnull().sum()
    if a == 21:
        paises_nulos.append(i)
print(paises_nulos)
```

```python
#Vemos que paises tienen datos faltantes
print("Paises con datos nulos")
paises_nulos=[]
for i in paises:
    mask = tbl_contaminacion_aire["id_pais"] == i
    df_filtrada = tbl_contaminacion_aire[mask]
    a = df_filtrada.iloc[:,2].isnull().sum()
    if a > 0:
        paises_nulos.append(i)
print(paises_nulos)
```

```python
#Cargamos la tabla con otro formato para tener una mejor visulizacion
wb.db= 2
tbl_contaminacion_aire2 = wb.data.DataFrame('EN.ATM.PM25.MC.M3', paises, time=range(2000, 2021), labels=True, columns="economy").reset_index()
#tbl_camas_hospitales.drop(["time","Country"], axis=1, inplace=True)
#tbl_camas_hospitales.columns=["Pais_ID","Year","Camas"]
tbl_contaminacion_aire2
```

INTERPOLAREMOS LOS DATOS FALTANTES

```python
#Todos los paises tienen datos nulos. Vamos a interpolar en ambas direcciones
for i in paises:
    temp = tbl_contaminacion_aire[tbl_contaminacion_aire['id_pais'] == i]
    temp = temp.interpolate(method="linear",limit_direction="backward")
    tbl_contaminacion_aire[tbl_contaminacion_aire['id_pais'] == i] = temp
```

```python
#CORROBORAMOS VALORES NULOS
tbl_contaminacion_aire.isnull().sum()
```
